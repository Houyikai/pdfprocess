```python
import fitz
import re

def read_pdf_with_spaces(filepath):
    doc = fitz.open(filepath)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# 判断text是否包含换行符
def has_newline(text):
    return '\n' in text

# 读取PDF文件并保留空格
filepaths = ["中华人民共和国数据安全法.pdf",'中华人民共和国网络安全法.pdf','中华人民共和国个人信息保护法.pdf']
paragraphs = []
for filepath in filepaths:
    paragraphs += re.split(r'\n|\s\s', read_pdf_with_spaces(filepath))
for line in paragraphs:
    print(line,'**')
```
